Prepare data for modelling

````python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression # type: ignore
from sklearn.ensemble import RandomForestClassifier
````

````python
train_df, test_df = train_test_split(df, test_size=0.1, random_state=42)
  # Create, random train and test set

 # Drop labels
train_df_labels = train_df["Outcome"].copy()
train_df= train_df.drop("Outcome", axis=1)
````

````python
#Transformation Pipelines

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler                       # This method makes normalization to our features

my_pipeline = Pipeline([('std_scaler', StandardScaler()), ])

train_prepared =my_pipeline.fit_transform(train_df)
````




Predictions

Make some predictions
Logistic Regression, RandomForestClassifier

````python

random_clf = RandomForestClassifier(random_state=42)
random_clf.fit(train_prepared, train_df_labels)
````

````python





logs_clf = LogisticRegression(random_state=42)
logs_clf.fit(train_prepared, train_df_labels)

````

````python
from sklearn.model_selection import cross_val_score

print("Random Forest Classifier ****\n", cross_val_score(random_clf, train_prepared, train_df_labels, cv= 3, scoring='accuracy'))
print("Logistic Regression: *** \n", cross_val_score(logs_clf, train_prepared, train_df_labels, cv= 3, scoring='accuracy'))
````

Predictions Accuracy

````python
prediction = random_clf.predict(train_prepared)
print("Random Forest Classifier Accuracy is \n: ", random_clf.score(train_prepared, train_df_labels))
````

````python
prediction = logs_clf.predict(train_prepared)
print("Logistics Regression Accuracy is \n: ", logs_clf.score(train_prepared, train_df_labels))
````

````python
from sklearn.model_selection import cross_val_predict
y_train_pred = cross_val_predict(logs_clf, train_prepared, train_df_labels, cv=3)
````

Let's look at the Confusion matrix

````python



from sklearn.metrics import confusion_matrix

confusion_matrix(train_df_labels, y_train_pred)

````

Precision = TP/(TP+FP) 392/(392+58)
Recall = TP/(TP+FN) 392/(392+109)
F1 Score = 2 x (precision x recall )/ (precision+recall)

````python
from sklearn.metrics import precision_score, recall_score, f1_score

print('Precision Score:',precision_score(train_df_labels, y_train_pred))
print('Recall Score:',recall_score(train_df_labels, y_train_pred))
print('F1 Score:',f1_score(train_df_labels, y_train_pred))
````
